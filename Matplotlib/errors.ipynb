```python
import matplotlib.pyplot as plt

plt.style.use("seaborn-whitegrid")
import numpy as np

x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt=".k")
```

```python
#errorbar's basic options
plt.errorbar(
    x,
    y,
    yerr=dy,
    fmt="o",
    color="black",
    ecolor="lightgray",
    elinewidth=3,
    capsize=0,
)

```
