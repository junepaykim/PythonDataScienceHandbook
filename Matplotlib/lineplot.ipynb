```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-whitegrid")

fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.cos(x))
plt.plot(x, np.tanh(x))
```

```python
# simple color name
plt.plot(x, np.sin(x - 0), color="blue")

# rgbcmyk
plt.plot(x, np.sin(x - 1), color="g")

# 0~1 by gray scale
plt.plot(x, np.sin(x - 2), color="0.75")

# RRGGBB
plt.plot(x, np.sin(x - 3), color="#FFDD44")

# RGB by 0~1
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))

# below is HTML color names
plt.plot(x, np.sin(x - 5), color="chartreuse")
```

```python
plt.plot(x, x + 0, linestyle="solid")
plt.plot(x, x + 1, linestyle="dashed")
plt.plot(x, x + 2, linestyle="dashdot")
plt.plot(x, x + 3, linestyle="dotted")

# For short, you can use the following codes
plt.plot(x, x + 4, linestyle="-")
plt.plot(x, x + 5, linestyle="--")
plt.plot(x, x + 6, linestyle="-.")
plt.plot(x, x + 7, linestyle=":")
```

```python
# line style and color by one argument with no name specified (non keyword argument)
plt.plot(x, x + 0, "-g")
plt.plot(x, x + 1, "--c")
plt.plot(x, x + 2, "-.k")
plt.plot(x, x + 3, ":r")
```

```python
mytemp = np.linspace(0, 10, 30)
plt.plot(mytemp, np.sin(mytemp))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

```python
plt.plot(x, np.sin(x))
# be careful not to call plt.axes
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis("tight")
```

```python
plt.plot(x, np.sin(x))
plt.axis("tight")
```

```python
plt.plot(x, np.sin(x))
plt.axis("equal")
```

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("my X")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), "-g", label="sin(x)")
plt.plot(x, np.cos(x), ":b", label="cos(x)")
plt.axis("equal")
plt.legend()
```

```python
#sometimes it is better to set all values once by ax.set function rather than set individually by set_xlabel, set_ylabel,set_xlim... etc
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(
    xlim=(0, 10),
    ylim=(-2, 2),
    xlabel="x",
    ylabel="sin(x)",
    title="A Simple Plot",
)
```
