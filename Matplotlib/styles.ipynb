```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = [1, 2, 3]
fig = plt.figure()
plt.plot(x, np.sin(x), "-")
plt.plot(x, np.cos(x), "--")
```
